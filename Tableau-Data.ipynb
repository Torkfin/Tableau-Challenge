```python
# Citibike Project
import pandas as pd
import datetime
import numpy as np
import os
```

```python
# Get NYC Citibike tripdata csv files from 2021
filepath_jan = os.path.join("Resources" ,"JC-202101-citibike-tripdata.csv")
filepath_feb = os.path.join("Resources" ,"JC-202102-citibike-tripdata.csv")
filepath_mar = os.path.join("Resources" ,"JC-202103-citibike-tripdata.csv")

nycbike_2021jan = pd.read_csv(filepath_jan)
nycbike_2021feb = pd.read_csv(filepath_feb)
nycbike_2021mar = pd.read_csv(filepath_mar)
```

```python
nycbike_2021jan.head(10)
```

```python
nycbike_2021 = pd.concat([nycbike_2021jan,nycbike_2021feb,nycbike_2021mar])
```

```python
# Output concatenated 2018 July-Aug citibike tripdata to CSV file 
filepath_2021 = os.path.join("Resources" ,"JC-2021-citibike-tripdata.csv")

nycbike_2021.to_csv(filepath_2021, index=False)

```
